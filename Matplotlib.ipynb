# Exercises

Use the data saved at ```./data/company_sales_data.csv``` to use for all exercises.

### Exercise 1. 

Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd
````

````python
# Exercise 1

import pandas as pd

company_sales = pd.read_csv("data/company_sales_data.csv")
# company_sales.describe()
company_sales
````

````python
x = company_sales["month_number"]
y = company_sales["total_profit"]
# line chart
plt.plot(x, y, linewidth = 3)

plt.title('Total Profit of All Months',fontsize = 13, fontweight = "bold")
plt.xlabel('Month', fontsize = 12, fontweight = "bold")
plt.ylabel('Total Profit (USD)', fontsize = 12, fontweight = "bold")

plt.tight_layout()
plt.show()
````

### Exercise 2. 

Get Total profit of all months and show line plot with the following Style properties

Generated line plot must include following Style properties:

1. Line Style dotted and Line-color should be red
2. Show legend at the lower right location.
3. X label name = Month Number
4. Y label name = Sold units number
5. Add a circle marker.
6. Line marker color as read
7. Line width should be 3

````python
x = company_sales["month_number"]
y = company_sales["total_profit"]
# line chart
plt.plot(x, y, linestyle = "dotted", color = "red", linewidth = 3,
        marker = "o", markeredgecolor = "red", label = "Total Profit Over Time")

plt.legend(loc = "lower right")
plt.xlim([1,12])
plt.title('Total Profit of All Months',fontsize = 13, fontweight = "bold")
plt.xlabel('Month Number', fontsize = 12, fontweight = "bold")
plt.ylabel('Sold units number', fontsize = 12, fontweight = "bold")
plt.tight_layout()
plt.show()
````

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
plt.rcParams["figure.figsize"] = (8,6)
plt.plot(x, company_sales["facecream"], linestyle = "--", color = "#AB82FF", linewidth = 2,
        label = "Facecream")

plt.plot(x, company_sales["facewash"], linestyle = "--", color = "#5a7d9a", linewidth = 2,
        label = "Facewash")

plt.plot(x, company_sales["toothpaste"], linestyle = "--", color = "#adad3b", linewidth = 2,
        marker = "o", markeredgecolor = "#adad3b", label = "Toothpaste")

plt.plot(x, company_sales["bathingsoap"], linestyle = "-", color = "#444444", linewidth = 2,
        marker = "v", markeredgecolor = "#444444", label = "bathingsoap")
plt.plot(x, company_sales["shampoo"], linestyle = "--", color = "#FFD700", linewidth = 2,
        marker = "*", markeredgecolor = "blue", label = "shampoo")
plt.plot(x, company_sales["moisturizer"], linestyle = "dotted", color = "#FF3030", linewidth = 2,
        marker = "x", markeredgecolor = "#FF3030", label = "moisturizer")

plt.legend(loc='upper left', bbox_to_anchor=(0.02, 1), fontsize = 12)
plt.xlim([1,12])
plt.title('Total Profit of All Months',fontsize = 13, fontweight = "bold")
plt.xlabel('Month Number', fontsize = 12, fontweight = "bold")
plt.ylabel('Sold units number', fontsize = 12, fontweight = "bold")
plt.tight_layout()
plt.show()
````

### Exercise 4.

Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should “–“.

````python
plt.rcParams["figure.figsize"] = (7,5)
plt.scatter(company_sales["month_number"], company_sales["toothpaste"])
plt.title('Toothpaste Sales',fontsize = 13, fontweight = "bold")
plt.xlabel('Month Number', fontsize = 12, fontweight = "bold")
plt.ylabel('Sold units number', fontsize = 12, fontweight = "bold")
plt.grid(axis = "both", linestyle = "-")
plt.tight_layout()
plt.show()
````

### Exercise 5.

Read face cream and face wash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

````python
# Exercise 5
plt.rcParams["figure.figsize"] = (8,6)
width = 0.25
plt.bar(company_sales["month_number"], company_sales["facewash"], width = width, color = "#C67171", label = "Facewash")
plt.bar(company_sales["month_number"]-width, company_sales["facecream"], width = width, color = "#7171C6", label = "Facecream")
plt.legend()
plt.title('Total Profit of All Months',fontsize = 13, fontweight = "bold")
plt.xlabel('Month Number', fontsize = 12, fontweight = "bold")
plt.ylabel('Sold units number', fontsize = 12, fontweight = "bold")
plt.tight_layout()
plt.show()
````
