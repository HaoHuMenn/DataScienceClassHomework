```python
df["Date"] = pd.to_datetime(df["Date"], format = "%Y-%m-%d")
df["Date"].shape[0]
i = 0
time_delta_lst = []
for i in range(df["Date"].shape[0]):
    if i == 0:
        continue
    else:
        time_delta = df["Date"][i]- df["Date"][i-1]
        time_delta_lst.append(time_delta)
#         print(df["Date"][i],df["Date"][i-1])
#         print(time_delta)

time_delta_lst
# mostly 7 days, dates in different year are not continuous, between 2014-11-25 and 2014-11-19 are 6 days. 
```

### Step 8.  Change the frequency to monthly, sum the values and assign it to monthly.

```python
monthly = df.resample("M").sum()
monthly
```

### Step 9. You will notice that it filled the dataFrame with months that don't have any data with NaN. Let's drop these rows.

```python
# change all the 0 to nan
monthly = monthly.replace(0, np.nan)
# drop nan on rows, when all values of a row is missing
monthly.dropna(axis = "index", how = "all", inplace = True)
```
