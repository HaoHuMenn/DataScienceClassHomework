### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import datetime
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data)

### Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python
# read data
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data"
# \s+ one or more white space characters
data = pd.read_csv(url, sep = "\s+", parse_dates = [[0,1,2]])
data.tail()
```

### Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
def fix(date):
    if date.year > 1978: # max year of this data sheet is 1978
        year = date.year - 100
    else:
        year = date.year
    # make a datetime format data
    
    d = datetime.date(year, date.month, date.day)
    return d
data["Yr_Mo_Dy"] = data["Yr_Mo_Dy"].apply(fix)
data.head()
```

### Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

```python
# dtype of Yr_Mo_Dy becomes object(datetime) after apply fix
# change Yr_Mo_Dy to datetime64[ns]
data["Yr_Mo_Dy"] = data["Yr_Mo_Dy"].astype("datetime64[ns]")

data.set_index("Yr_Mo_Dy", inplace = True)
```

### Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.

#### A different set of numbers for each day.

```python
day_stats = pd.DataFrame(index = data.index)
day_stats["min"] = data.min(axis = 1)
day_stats["max"]  = data.max(axis = 1)
day_stats["mean"] = data.mean(axis = 1)
day_stats["standard_deviations"] = data.std(axis = 1)
day_stats
```

### Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

```python
# convert Jan 2 1961 to datetime.date
d = datetime.date(1961, 1, 2)

# see which weekday Jan 2, 1961 is. it is 1, Monday
d.isoweekday()

# so week starts on Jan 2, 1961 means starts on Mon
# resample weekly starts on Mon, calculate min, max, mean, std
# Anchored offsets at: https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#anchored-offsets
weekly = data.resample("W-MON").agg(["min", "max", "mean", "std"])

#grab first 52 DatetimeIndex
first_52w_idx = weekly.index[0:52]

# first 52 weeks of result
weekly.loc[first_52w_idx, :]
```
